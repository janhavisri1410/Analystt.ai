```python
#Importing the libraries

import pandas as pd
import pdfquery
```

```python
#loading the pdf file

pdf = pdfquery.PDFQuery('File.pdf')
pdf.load()
pdf.tree.write('pdfXML.txt', pretty_print = True)   #Dumping the xml file
```

### Getting the page count ###

```python
pagecount = pdf.doc.catalog['Pages'].resolve()['Count']
print(pagecount)
```

### Creating 2 Dataframes for alternate side of pages ###

```python
def pdfscrape_1(pdf):
    # Extract each relevant information individually
    name             = pdf.pq('LTTextLineHorizontal:overlaps_bbox("48.693, 462.323, 205.479, 474.323")').text()
    phone            = pdf.pq('LTTextLineHorizontal:overlaps_bbox("154.823, 425.174, 265.595, 434.174")').text()
    mobile           = pdf.pq('LTTextLineHorizontal:overlaps_bbox("154.823, 436.007, 222.278, 445.007")').text()
    #email            = pdf.pq('LTTextLineHorizontal:overlaps_bbox("154.823, 370.828, 259.691, 379.828")').text()#
    #office_address   = pdf.pq('LTTextLineHorizontal:overlaps_bbox("48,  543,  260, 577")').text()
    #residence_address= pdf.pq('LTTextLineHorizontal:overlaps_bbox("370, 662,  430, 672")').text()
# Combined all relevant information into single observation
    page = pd.DataFrame({
                         'employer_name': name,
                         'employer_phone': phone,
                         'mobile': mobile
                         #'email': email
                         #'office_address': office_address,
                         #'residence_address': residence_address
                       }, index=[0])
    return(page)
```

```python
master_1 = pd.DataFrame()
for p in range(pagecount):
    pdf.load(p)
    page = pdfscrape_1(pdf)
    master_1 = master_1.append(page, ignore_index=True)

master_1
```

### Getting the addresses ###

```python
def get_address(pdf):
    keyword = pdf.pq('LTTextLineHorizontal:contains("{}")'.format("Physical","Office"))[0]
    x0 = float(keyword.get('x0',0)) +30
    y0 = float(keyword.get('y0',0)) -60
    x1 = float(keyword.get('x1',0)) +30
    y1 = float(keyword.get('y1',0)) -20
    address = pdf.pq('LTTextLineHorizontal:overlaps_bbox("%s, %s, %s, %s")' % (x0, y0, x1, y1)).text()
    print(address)


address = pd.DataFrame()
for p in range(pagecount):
    pdf.load(p)
    page = get_address(pdf)
    address = address.append(page, ignore_index=True)

address.head()
```
